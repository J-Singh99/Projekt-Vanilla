```python
import pandas as pd
```

```python
df=pd.read_csv('God_Given_Gift.csv')
df
```

```python
pd=[]
for rows in df['Job_Summary']:
    pd.append(rows)
print(pd,end="")
```

```python
list1=[]
for explode in pd:
    list1.append(explode.split("\n"))
list1
```

```python

for ele in list1:
    for i in range(5):
        ele.remove(ele[0])

    for i in range(10):
        ele.remove(ele[-1])

list1
```
