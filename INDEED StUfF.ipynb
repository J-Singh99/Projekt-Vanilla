```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
shiz = requests.get('https://www.indeed.co.in/software-testing-jobs-in-Defence-Colony,-Delhi,-Delhi')
```

```python
soup = BeautifulSoup(shiz.text,'html.parser')
```

```python
count = 0
for i in soup.find_all(class_= 'jobsearch-SerpJobCard unifiedRow row result'):
    print('************************')
    print(i.find(attrs={'data-tn-element' : 'jobTitle'})['title'])
    print(i.find(class_ = 'company').text)
    print(i.find(class_ = 'location accessible-contrast-color-location').text)
    if i.find(class_ = 'salaryText') == None:
        continue
    print(i.find(class_ = 'salaryText').text)
```
